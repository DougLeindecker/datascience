```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
arq = ('/home/douglas/Downloads/credit_risk_dataset.csv')
dados = pd.read_csv(arq)
```

```python
#Checando dados nulos
print((dados.isnull().sum()))
print(' ')
print(f'A média do loan_int_rate = {dados.loan_int_rate.mean()}')
print(' ')
print(f'A média de person_emp_length =  {dados.person_emp_length.mean()}')
print(' ')
dados2 = dados
```

```python
print(f'Valores faltantes antes da alteração = {dados2.loan_int_rate.isnull().sum()}')
# Preenchedo os valores faltantes com a média dos dados para loan_int_rate
dados2.loan_int_rate.fillna(dados2.loan_int_rate.mean(), inplace=True)
print(' ')
print(f'Valores faltantes depois da alteração = {dados2.loan_int_rate.isnull().sum()}')
print(' ')
print(f'Valores faltantes antes da alteração = {dados2.person_emp_length.isnull().sum()}')
```

```python
# Preenchedo os valores faltantes com a média dos dados para person_emp_length
print(' ')
dados2.person_emp_length.fillna(dados2.person_emp_length.mean(), inplace=True)
print(f'Valores faltantes depois da alteração = {dados2.person_emp_length.isnull().sum()}')
```

```python
dados3 = dados2.drop(dados2[dados2['person_age']>99].index)
```

```python
plt.figure(figsize=(20,10))
plt.title('Intenção de empréstimo por idade')
plt.ylabel('Idade')
plt.xlabel('Intenção do empréstimo')
plt.scatter(y=dados3.person_age,x=dados3.loan_intent)
```

```python
plt.figure(figsize=(20,10))
plt.title('Frequência de idades')
plt.ylabel('Frequência')
plt.xlabel('Idade')
plt.hist(dados3.person_age,edgecolor='black')
```

```python
plt.figure(figsize=(20,10))
plt.title('Idade X emprestimo')
plt.ylabel('Idade')
plt.xlabel('Motivo de emprestimo')
sns.barplot(dados3.loan_intent,dados3.person_age)
```
